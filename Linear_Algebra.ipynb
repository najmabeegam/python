```python
import numpy as np
```

```python
a = np.array([[10, 15]])
b = np.array([[20, 9]])
print(a)
print(b)
print()

a + b
a.T + b.T #or (a + b).T (for transpose)
```

```python
#Substracting 2 vectors

a - b
#(a + b).T (for transpose)
```

```python
#Scalar Multiplication

a = 2
u = np.array([[30, 15]])

a * u
```

```python
(a * u).T       #transpose
```

```python
# Dot product of 2 vectors
# For 2-D vectors, it is equivalent to matrix multiplication

u = np.array([10, 20])
v = np.array([25, 6])

print(10 * 25 + 20 * 6)

np.dot(u,v)
```

```python
# Matrix Addition

a = np.array([[12, 10], [4, 6], [40,31],])
b = np.array([[2, 8], [46, 63], [50,10],])


print(a, end=" = a\n\n")
print(b, end=" = b\n\n")

a + b
```

```python
# Matrix Substraction
a - b
```

```python
# Matrix scalar multiplication

a = 4
b = np.array([[2, 5], [3, 7], [8, 6],])

a * b
```

```python
# Matrix and vector multiplication

a = [[5, 10], [15,20]]
b = [10, 12]

np.dot(a,b)
```

```python
# Matrix to Matrix multiplication

a = [[5, 10], [15, 20], [25, 30]]
b = [[3, 6], [9, 5]]

np.dot(a, b)
```

```python
# Transpose a matrix

a = np.array([[5, 10], [15, 20], [25,30]])
print(a, end=" = a\n\n")
print(a.T, end=" =aT")
```

```python
# Identity matrix  (creates a diagonal of 1)
I1 = np.eye(2)
I2 = np.eye(3)
print(I1, end=" = I1\n\n")
print(I2, end=" = I2\n\n")
```
